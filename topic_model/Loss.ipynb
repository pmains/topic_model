```python
import pandas as pd
import altair as alt
```

```python
loss_df = pd.read_csv("loss.csv")
loss_df.head()
```

```python
run_df = pd.read_csv("runs.csv")
run_df.head()
```

```python
len(loss_df), len(run_df)
```

```python
alt.Chart(loss_df[:5000]).mark_line().encode(
    x=alt.X("epoch"),
    y=alt.Y("loss"),
    color=alt.Color("run", legend=None),
    tooltip=['run']
).properties(
    height=600,
    width=800
)
```

```python
run_code = '4dNIfvWQ'
loss_df[loss_df.run == run_code].head()
```

```python
run_df[run_df.run_code == run_code]
```

```python
batch = 4
start = start = batch * 5000
end = start + 5000

alt.Chart(loss_df[start:end]).mark_line().encode(
    x=alt.X("epoch"),
    y=alt.Y("loss"),
    color=alt.Color("run", legend=None),
    tooltip=['run']
).properties(
    height=600,
    width=800
)
```
